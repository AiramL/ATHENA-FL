```python
import numpy as np
import tensorflow as tf
from pickle import load
from matplotlib import pyplot
```

```python
x_train2 = np.asarray(load(open('../../../../datasets/CIFAR-10/Non-IID-distribution/train/class6Train','rb')))
x_trainL = np.asarray(load(open('../../../../datasets/CIFAR-10/Non-IID-distribution/train/class6TrainLabel','rb')))
pyplot.imshow(x_train2[1000])
```

```python
from numpy import expand_dims
from keras.preprocessing.image import ImageDataGenerator

def generate_data(image_array):
    sample_as_array = expand_dims(image_array,0)
    datagen = ImageDataGenerator(horizontal_flip=True)
    it = datagen.flow(sample_as_array, batch_size=1)
    batch = it.next()
    new_image = batch[0].astype('uint8')
    return new_image
```

```python

dataset = []
labels = []

for (data, label) in zip(x_train2, x_trainL):
    dataset.append(generate_data(data))
    labels.append(label)
    
```
